```python
import numpy as np
import matplotlib.pyplot as plt

cmap = plt.cm.viridis.copy()
cmap.set_bad(color='black')
```

```python
n = 20
H = np.triu(np.random.randn(n, n), -1)
HH = H @ H
HHH = HH @ H

plt.imshow(H, cmap, vmin=-1, vmax=1)
plt.show()
plt.imshow(HH, cmap, vmin=-1, vmax=1)
plt.show()
plt.imshow(HHH, cmap, vmin=-1, vmax=1)
plt.show()
```
